# Algerian Forest Fires Dataset

Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012. The dataset includes 11 attribues and 1 output attribue (class) The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
data = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
```

```python
data.head()
```

```python
data.info()
```

# cleaning the data

```python
data.isnull().sum()
```

```python
data[data.isnull().any(axis=1)]
```

As we saw in the beginning there are 2 dataset or regions of data in this single file 


```python
data.loc[:122,"Region"]=0
data.loc[122:,"Region"]=1
df = data
```

```python
df.iloc[[122]]
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

### now we meerged the both data set and classified it as region 1 and region 0 and aslo we cleaned the side headings and also the null values 

```python
df.columns
```

```python
## we can see spaces in the column names which might cause headace further so let's clean that

df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

#### all the columns here are in the object data type and needs type conversion so let's do it

```python
df[['month','day','year','Temperature','RH','Ws',"Region"]] = df[['month','day','year','Temperature','RH','Ws','Region']].astype(int)
```

```python
df.info()
```

Changing the other columns to float datatype

```python
objects=[features for features in df.columns if df[features].dtypes=='object']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('cleaned_algerian_dataset.csv',index=False)

##saving this inorder to build a model as it's cleaned and ready
```

# Exploratory data analysis

```python
## categories in classes
df['Classes'].value_counts()
```

```python
## Encoding of the categories in classes
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df['Classes'].value_counts()
```

```python
data = df.groupby('month')['Classes'].sum()

plt.figure(figsize=(8, 8))
plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140)

plt.title('Classes Distribution by Month')
plt.show()
```

Here we can see that in th August month the fire chances are high

```python
data = df.groupby('day')['Classes'].sum()

plt.figure(figsize=(8, 8))
plt.pie(data, labels=data.index, autopct='%1.1f%%', startangle=140)

plt.title('Classes Distribution by Day')
plt.show()
```

```python
## drop day,month and year
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
## Plot desnity plot for all features
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## Percentage for Pie Chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
# plotting piechart
classlabels=["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```
